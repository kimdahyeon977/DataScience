```python
frequency=list(map(lambda x:float(x[:-5]),age_data['전체표준오차'].values))
labels=age_data['연령']
plt.rc('font', family='Malgun Gothic') #한글로
plt.pie(frequency, labels=labels, autopct='%.1f%%')
plt.show()
#40-49가 가장 평균근처에 밀집되어있고 60-69가 평균에서 가장 멀리 떨어져있다.
```
